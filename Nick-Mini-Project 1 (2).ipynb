# Exploring Probability Distributions (Poisson Distribution) 

A Poisson distribution is a discrete probability distribution. The Poisson distribution has only one parameter, λ (lambda), which is the mean number of events. 

It's named after French mathematician Simeon Denis Poisson. It's used to model the number of events that occur in a fixed interval of time or space, given the average rate of occurrence, assuming that the events happen independently and at a constant rate.
It deals with discrete random variables, meaning the number of events can only take on non-negative integer values (0, 1, 2, 3,…). Each event is considered to be independent of others and they are assumed to occur at a constant average rate (λ) over the given interval.

What this basically means is that when we know the average number of times that an event will happen, we can use Poisson to calculate how likely other numbers deviate from this average.


### Characteristics 

Since the Poisson distribution measures the frequency of events under the assumption of statistical randomness, the agreement of the expected distribution with the actual data suggests that the actual data was indeed due to randomness. If the actual data resulted in many more deaths than expected, an alternate explanation should be sought (e.g. inadequate training, a clever and subtle enemy plot, etc.).
The Poisson distribution is useful in determining the probability that a certain number of events occur over a given time period. For example, if an office averages 12 calls per hour, they can calculate that the probability of receiving at least 20 calls in an hour is
* $\sum_{k=20}^{\infty} \frac{12^k e^{-12}}{k!} \approx 2.12\%$

Which means they can generally feel comfortable keeping only enough staff on hand to handle 20 calls. 
The list of potential applications is very long. Some of the examples are listed below: 
* the number of mutations on a given strand of DNA per time unit
* the number of bankruptcies that are filed in a month
* the number of arrivals at a car wash in one hour
* the number of network failures per day
* the number of file server virus infection at a data centre during a 24-hour period

### Assumptions and validity
The Poisson distribution is an appropriate model if the following assumptions are true:
* k is the number of times an event occurs in an interval and k can take values 0, 1, 2, ... .
* The occurrence of one event does not affect the probability that a second event will occur. That is, events occur independently.
* The average rate at which events occur is independent of any occurrences. For simplicity, this is usually assumed to be constant, but may in practice vary with time.
* Two events cannot occur at exactly the same instant; instead, at each very small sub-interval, either exactly one event occurs, or no event occurs.
If these conditions are true, then k is a Poisson random variable, and the distribution of k is a Poisson distribution.!


As a keen soccer fan (or football depending on where you grew up) I'm going to try and apply the above principles to explain statistics we commonly see in real world matches. 

The Poisson distribution is particularly useful for this because 
* It models the number of events (goals) in a fixed interval of time (a match).
* It assumes events occur with a known average rate and independently of each other.
* In soccer, goals are relatively rare events that occur at a somewhat constant rate, making Poisson a good fit.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_soccer_match(avg_goals_per_team, num_simulations):
    """
    Simulate goals scored by two teams in a soccer match using Poisson distribution.
    
    Parameters:
    avg_goals_per_team (float): Average number of goals scored by a team in a match
    num_simulations (int): Number of matches to simulate
    
    Returns:
    tuple: Two arrays representing goals scored by each team in simulated matches
    """
    team1_goals = np.random.poisson(avg_goals_per_team, num_simulations)
    team2_goals = np.random.poisson(avg_goals_per_team, num_simulations)
    return team1_goals, team2_goals

def plot_results(team1_goals, team2_goals):
    """
    Plot the distribution of goals scored by both teams.
    
    Parameters:
    team1_goals (array): Goals scored by team 1 in simulated matches
    team2_goals (array): Goals scored by team 2 in simulated matches
    """
    plt.figure(figsize=(10, 6))
    plt.hist([team1_goals, team2_goals], label=['Team 1', 'Team 2'], bins=range(max(np.max(team1_goals), np.max(team2_goals))+2))
    plt.xlabel('Number of Goals')
    plt.ylabel('Frequency')
    plt.title('Distribution of Goals Scored in Simulated Soccer Matches')
    plt.legend()
    plt.show()

# Set parameters
avg_goals_per_team = 1.5  # Average goals per team per match
num_simulations = 10000   # Number of matches to simulate

# Run simulation
team1_goals, team2_goals = simulate_soccer_match(avg_goals_per_team, num_simulations)

# Calculate and print some statistics
print(f"Average goals for Team 1: {np.mean(team1_goals):.2f}")
print(f"Average goals for Team 2: {np.mean(team2_goals):.2f}")
print(f"Maximum goals scored in a single match: {max(np.max(team1_goals), np.max(team2_goals))}")

# Plot the results
plot_results(team1_goals, team2_goals)
```

If we change the parameters (average goals a game) of the above code we can begin to see the affect on the distibution's shape and properties. I note that the main differences were: 

Shape of the Distribution:
 * As the average goals increase, the peak of the distribution shifts to the right (begins to resemble a normal distribution) 
 * For lower averages (0.5, 1.0), the distribution is highly skewed to the left, with the highest probability at 0 goals.
 
Mean and Variance:
 * In a Poisson distribution, the mean and variance are theoretically equal to the parameter (average goals in my example). The simulations match this closely (see mean/variance in top right corner of graphs)
 
Spread of the Distribution:
 * As the average increases, the distribution spreads out more. Also the standard deviation increases

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_soccer_match(avg_goals_per_team, num_simulations):
    """
    Simulate goals scored by two teams in a soccer match using Poisson distribution.
    
    Parameters:
    avg_goals_per_team (float): Average number of goals scored by a team in a match
    num_simulations (int): Number of matches to simulate
    
    Returns:
    tuple: Two arrays representing goals scored by each team in simulated matches
    """
    team1_goals = np.random.poisson(avg_goals_per_team, num_simulations)
    team2_goals = np.random.poisson(avg_goals_per_team, num_simulations)
    return team1_goals, team2_goals

def plot_results(team_goals, avg_goals, ax):
    """
    Plot the distribution of goals scored on a given subplot.
    
    Parameters:
    team_goals (array): Goals scored in simulated matches
    avg_goals (float): Average goals used for this simulation
    ax (matplotlib.axes.Axes): Subplot to plot on
    """
    ax.hist(team_goals, bins=range(max(team_goals)+2), density=True, alpha=0.7)
    ax.set_xlabel('Number of Goals')
    ax.set_ylabel('Probability')
    ax.set_title(f'Goal Distribution (Avg: {avg_goals:.1f})')
    ax.set_xlim(0, 10)  # Limit x-axis for better comparison
    
    # Calculate and display mean and variance
    mean = np.mean(team_goals)
    variance = np.var(team_goals)
    ax.text(0.95, 0.95, f'Mean: {mean:.2f}\nVariance: {variance:.2f}', 
            transform=ax.transAxes, verticalalignment='top', horizontalalignment='right')

# Set parameters
avg_goals_list = [0.5, 1.0, 2.0, 3.0]  # Different average goals to simulate
num_simulations = 10000   # Number of matches to simulate for each average

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 12))
axs = axs.ravel()  # Flatten axs array for easy indexing

# Simulate and plot for each average goal value
for i, avg_goals in enumerate(avg_goals_list):
    team_goals, _ = simulate_soccer_match(avg_goals, num_simulations)
    plot_results(team_goals, avg_goals, axs[i])

plt.tight_layout()
plt.show()

# Print additional statistics
print("Additional Statistics:")
for avg_goals in avg_goals_list:
    team_goals, _ = simulate_soccer_match(avg_goals, num_simulations)
    print(f"\nFor average goals = {avg_goals:.1f}")
    print(f"  Mean: {np.mean(team_goals):.2f}")
    print(f"  Variance: {np.var(team_goals):.2f}")
    print(f"  Std Dev: {np.std(team_goals):.2f}")
    print(f"  % of 0-goal matches: {np.mean(team_goals == 0) * 100:.2f}%")
    print(f"  % of matches with 3+ goals: {np.mean(team_goals >= 3) * 100:.2f}%")
```

## Significance test for the Poisson distribution:

Hypotheses:
* Null Hypothesis (H0): The true mean (λ) of the Poisson distribution is equal to the expected rate (λ0).
* Alternative Hypothesis (H1): The true mean (λ) of the Poisson distribution is not equal to the expected rate (λ0).

In the code below I'll simulate 100 matches with an average of 1.5 goals per match and perform significance test to determine if this dataset is consistent with an expected rate of 1.5 goals per match.

The test calculates the test statistic and p-value, and determines whether to reject the null hypothesis based on a significance level of 0.05. A small p-value suggest evidence against the null hypothesis and large p-value suggests the data is consistent with the null hypothesis

```python
from scipy import stats

def poisson_significance_test(observed_data, expected_rate, alpha=0.05):
    """
    Perform a significance test for a Poisson distribution.
    
    Parameters:
    observed_data (array-like): Observed count data
    expected_rate (float): Expected rate (mean) of the Poisson distribution
    alpha (float): Significance level (default 0.05)
    
    Returns:
    tuple: Test statistic, p-value, and boolean indicating if null hypothesis is rejected
    """
    n = len(observed_data)
    observed_mean = np.mean(observed_data)
    
    # Calculate test statistic
    test_statistic = (observed_mean - expected_rate) / (expected_rate / np.sqrt(n))
    
    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(test_statistic)))
    
    # Determine if null hypothesis is rejected
    reject_null = p_value < alpha
    
    return test_statistic, p_value, reject_null

# Example usage
observed_goals = np.random.poisson(1.5, 100)  # Simulate 100 matches with average 1.5 goals
expected_rate = 1.5

test_statistic, p_value, reject_null = poisson_significance_test(observed_goals, expected_rate)

print(f"Test Statistic: {test_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Reject Null Hypothesis: {reject_null}")

# Interpretation
if reject_null:
    print("There is significant evidence to suggest that the true mean differs from the expected rate.")
else:
    print("There is not enough evidence to conclude that the true mean differs from the expected rate.")
```

# Conclusion 
A Poisson Process may be a model for a series of discrete event where the typical time between events is understood, but the precise timing of events needs to be random. The arrival of an occasion is independent of the event before (waiting time between events is memory less).

Having said this, the Poisson distribution has many applications to count data (telecommunication, science, finance and insurance, radioactivity and list goes on) because of it's ability to predict the probability of a given number of events based on a known average.
